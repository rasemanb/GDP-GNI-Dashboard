```python
import pandas as pd 
import json
from sqlalchemy import create_engine
import numpy as np
```

```python
csv_file = "./Resources/GDP.csv"
gdp_df = pd.read_csv(csv_file)
gdp_df
```

```python
data_df = gdp_df[['Country Name', 'Series Name', '2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]', '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]', '2020 [YR2020]', '2021 [YR2021]']].copy()
data_df
```

```python
data_df.replace('..', np.nan, inplace=True)

drop_df = data_df.dropna(how='any')
drop_df
```

```python
rename_df = drop_df.rename(columns = {"2012 [YR2012]":"2012",
                         "2013 [YR2013]":"2013", 
                         "2014 [YR2014]":"2014", 
                         "2015 [YR2015]":"2015", 
                         "2016 [YR2016]":"2016",
                         "2017 [YR2017]":"2017",
                         "2018 [YR2018]":"2018",
                         "2019 [YR2019]":"2019",
                         "2020 [YR2020]":"2020",
                         "2021 [YR2021]":"2021"})
rename_df
```

```python
clean_df = rename_df.loc[drop_df['Country Name'].isin(['United States', 'China', 'Japan', 'Germany', 'India', 'United Kingdom', 'France', 'Italy', 'Canada', 'Korea, Rep.'])]
clean_df
```
